Пункт 1. Вывести датасет для удобной работы.

```python
import pandas as pd
import numpy as np

games = pd.read_csv("games.csv")
games
```

Пункт 2. Создание линейных графиков - 3 штуки.

```python
import matplotlib.pyplot as plt
x = games["white_rating"].head()
y = games["turns"].head()
plt.plot(x,y, color ='red', marker = 'o', label = 'Линейный график рейтинга белых игроков')
plt.xlabel('Рейтинг белых игроков')
plt.ylabel('Количество ходов')
plt.grid()
plt.legend()
plt.show()
```

1. Данный график отображает отношение между количеством ходов игроков и их рейтингом. На этом графике можно заметить, что чем меньше ходов совершает игрок, тем  выше его рейтинг.

```python
x = games["black_rating"].head()
y = games["turns"].head()
plt.plot(x,y, color ='black', marker = 'o', label = 'Линейный график рейтинга черных игроков')
plt.xlabel('Рейтинг черных игроков')
plt.ylabel('Количество ходов')
plt.grid()
plt.legend()
plt.show()
```

2. Данный график отображает отношение между количеством ходов игроков и их рейтингом. На этом графике можно заметить, что чем меньше ходов совершает игрок, тем  выше его рейтинг. Черные игроки совершают меньше ходов чем белые и  их рейтиг выше.

```python
x = games["created_at"].head()
y = games["last_move_at"].head()
plt.plot(x,y, color ='green', marker = 'o',markersize = 10, label = 'Линейный график времени игры')
plt.xlabel('Время начала')
plt.ylabel('Время окончания')
plt.grid()
plt.legend()
plt.show()
```

3. На данном графике изображено соотношение времени начала игры и времени её окончания.

Пункт 3. Создание столбчатой диаграммы -  3 штуки.

```python

plt.figure(figsize=(14,14))
x = games["moves"][10:15]
y = games["increment_code"][10:15]
plt.bar (x, y, label = 'Соотношение статуса времени и всех ходов', color = "purple")
plt.xlabel('Все ходы в стандартной шахматной записи')
plt.ylabel('Приращение времени')
plt.title('Столбчатая диаграмма время/ходы')
plt.legend()
plt.show()
```

1. На данной столбчатой диаграмме идет сравнение все ходы в стандартной шахматной записи и приращение времени.

```python
plt.figure(figsize=(7,7))
x = games["turns"][10:15]
y = games["increment_code"][10:15]
plt.bar (x, y, label = 'Соотношение всех ходов и ходов в начальной фазе', color = "pink")
plt.xlabel('Количество ходов')
plt.ylabel('Количество ходов в начальной фазе')
plt.title('Столбчатая диаграмма ходы/ходы в начальной фазе')
plt.legend()
plt.show()
```

2. На данной столбчатой диаграмме идет сравнение все ходы в стандартной шахматной записи и количество ходов в начальной фазе.

```python
plt.figure(figsize=(12,12))
x = games["black_id"][10:15]
y = games["black_rating"][10:15]
plt.bar (x, y, label = 'Соотношение идентификатора и рейтинга', color = "blue")
plt.xlabel('Идентификатор игрока играющего за черные шахматы')
plt.ylabel('Рейтинг черных игроков')
plt.title('Столбчатая диаграмма идентификатор/рейтинг')
plt.legend()
plt.show()
```

3. На данной столбчатой диаграмме идет сравнение идентификатора игрока играющего за черные шахматы и его рейтинга.

Пункт 4. Круговая диаграмма - 2 штуки

```python
vals = games["white_rating"][10:15]
labels = games["white_id"][10:15]
plt.pie(vals, labels = labels, autopct ='%1.1f%%')
plt.title("Распределение идентификатора белого игрока и его рейтинга")
plt.show() 
```

1. На данной круговой диаграмме можно увидеть распределение идентификатора белого игрока и его рейтинга.

```python
vals = games["black_rating"][10:15]
labels = games["black_id"][10:15]
plt.pie(vals, labels = labels, autopct ='%1.1f%%')
plt.title("Распределение идентификатора черного игрока и его рейтинга")
plt.show() 
```

2. На данной круговой диаграмме можно увидеть распределение идентификатора черного игрока и его рейтинга.

Пункт 5. Диаграмма рассеяния - 2 штуки.

```python
plt.figure(figsize=(7,7))
x = games['moves'][10:15]
y = games['opening_ply'][10:15]
plt.title("Диаграмма рассеяния ходы в стандартной шахматной записи/ходы в начальной фазе")
plt.xlabel('Все ходы в стандартной шахматной записи')
plt.ylabel('Количество ходов в начальной фазе')
plt.scatter(x,y)
plt.show()
```

1. На данной диаграмме рассеяния можно увидеть все ходы в стандартной шахматной записи и количество ходов в начальной фазе.

```python
plt.figure(figsize=(7,7))
x = games['victory_status'][10:20]
y = games['winner'][10:20]
plt.title("Диаграмма рассеяния статус игры/победитель")
plt.xlabel('Статус игры')
plt.ylabel('Победитель')
plt.scatter(x,y)
plt.show()
```

2. На данной диаграмме рассеяния можно увидеть статус игры и победителя в этой игре.

Пункт 6.Гистограмма - 3 штуки.

```python
plt.figure(figsize=(7,7))
plt.hist(games['created_at'], bins = 30, color =["black"],
 edgecolor='pink',alpha=0.7, label ='Распределение времени игроков')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

1. На данной гистограмме можно рассмотреть распределение времени игроков по точным значениям.

```python
plt.figure(figsize=(7,7))
plt.hist(games['turns'], bins = 30, color =["pink"],
 edgecolor='black',alpha=0.7, label ='Распределение количество ходов')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

2. На данной гистограмме можно рассмотреть распределение количества ходов по точным значениям.

```python
plt.figure(figsize=(7,7))
plt.hist(games['winner'], bins = 30, color =["pink"],
 edgecolor='black',alpha=0.7, label ='Количество победителей')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

3. На данной гистограмме можно рассмотреть распределение количества победителей по точным значениям.Мы можем видеть цвета победителей и их колличество.
