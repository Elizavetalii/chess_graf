Пункт 1. Вывести датасет для удобной работы.

```python
import pandas as pd
import numpy as np

games = pd.read_csv("games.csv")
games
```

Пункт 2. Создание линейных графиков - 3 штуки.

```python
import matplotlib.pyplot as plt
x = games["white_rating"].head()
y = games["turns"].head()
plt.plot(x,y, color ='red', marker = 'o', label = 'Линейный график рейтинга белых игроков')
plt.xlabel('Рейтинг белых игроков')
plt.ylabel('Количество ходов')
plt.grid()
plt.legend()
plt.show()
```

1. Данный график отображает отношение между количеством ходов игроков и их рейтингом. На этом графике можно заметить, что чем меньше ходов совершает игрок, тем  выше его рейтинг. К примеру, если рейтинг игрока равен 1325, то за свою игру он сделал чуть меньше 20 ходов. 

```python
x = games["black_rating"].head()
y = games["turns"].head()
plt.plot(x,y, color ='black', marker = 'o', label = 'Линейный график рейтинга черных игроков')
plt.xlabel('Рейтинг черных игроков')
plt.ylabel('Количество ходов')
plt.grid()
plt.legend()
plt.show()
```

2. Данный график отображает отношение между количеством ходов игроков и их рейтингом. На этом графике можно заметить, что чем меньше ходов совершает игрок, тем  выше его рейтинг. Черные игроки совершают меньше ходов чем белые и  их рейтиг выше. К примеру, если рейтинг игрока равен 1150, то за свою игру он сделал около 10 ходов или даже меньше.

```python
x = games["created_at"].head()
y = games["last_move_at"].head()
plt.plot(x,y, color ='green', marker = 'o',markersize = 10, label = 'Линейный график времени игры')
plt.xlabel('Время начала')
plt.ylabel('Время окончания')
plt.grid()
plt.legend()
plt.show()
```

3. На данном графике изображено соотношение времени начала игры и времени её окончания. Можно заметить, что иногда время начала шахмантной партии совпадает с временем ее окончания, таков уж этот датасет. Например, когда время начала партии равно 1.30, то время окончания равно 1.30.

Пункт 3. Создание столбчатой диаграммы -  3 штуки.

```python

plt.figure(figsize=(14,14))
x = games["moves"][10:15]
y = games["increment_code"][10:15]
plt.bar (x, y, label = 'Соотношение статуса времени и всех ходов', color = "purple")
plt.xlabel('Все ходы в стандартной шахматной записи')
plt.ylabel('Приращение времени')
plt.title('Столбчатая диаграмма время/ходы')
plt.legend()
plt.show()
```

1. На данной столбчатой диаграмме идет сравнение все ходы в стандартной шахматной записи и приращение времени. Так как ходов в шахматной партии большое количество они наложились друг на друга в диаграмме, но мы можем четко рассмотреть тот факт, что чем больше наше приращение времени тем больше было совершено ходов за игру.

```python
plt.figure(figsize=(7,7))
x = games["turns"][10:15]
y = games["increment_code"][10:15]
plt.bar (x, y, label = 'Соотношение всех ходов и ходов в начальной фазе', color = "pink")
plt.xlabel('Количество ходов')
plt.ylabel('Количество ходов в начальной фазе')
plt.title('Столбчатая диаграмма ходы/ходы в начальной фазе')
plt.legend()
plt.show()
```

2. На данной столбчатой диаграмме идет сравнение все ходы в стандартной шахматной записи и количество ходов в начальной фазе. Например, чем меньше у игрока колличества общих ходов за игру, тем больше ходов в начальной фазе он совершил. Можно рассмотреть, что когда общее кол-во ходов равно чуть больше 30 тогда кол-во ходов в начальной фазе 15+15, а когда общее количество ходов равно 60, кол-во ходов  в начальной фазе 20+60.

```python
plt.figure(figsize=(12,12))
x = games["black_id"][10:15]
y = games["black_rating"][10:15]
plt.bar (x, y, label = 'Соотношение идентификатора и рейтинга', color = "blue")
plt.xlabel('Идентификатор игрока играющего за черные шахматы')
plt.ylabel('Рейтинг черных игроков')
plt.title('Столбчатая диаграмма идентификатор/рейтинг')
plt.legend()
plt.show()
```

3. На данной столбчатой диаграмме идет сравнение идентификатора игрока играющего за черные шахматы и его рейтинга. Здесь можно четко рассмотреть имя игрока и его рейтинг, например, у игрока с ником anaisaac рейтинг равен 1250.

Пункт 4. Круговая диаграмма - 2 штуки

```python
vals = games["white_rating"].head()
labels = games["white_id"].head()
plt.pie(vals, labels = labels, autopct ='%1.1f%%')
plt.title("Распределение идентификатора белого игрока и его рейтинга")
plt.show() 
```

1. На данной круговой диаграмме можно увидеть распределение идентификатора белого игрока и его рейтинга. Здесь в четком процентном соотношении можно рассмотреть рейтинг игроков и его имени. Например, игрок с ником a-00 по рейтингу а процентвом соотношении меньше всего 18,2%.

```python
vals = games["black_rating"][10:15]
labels = games["black_id"][10:15]
plt.pie(vals, labels = labels, autopct ='%1.1f%%')
plt.title("Распределение идентификатора черного игрока и его рейтинга")
plt.show() 
```

2. На данной круговой диаграмме можно увидеть распределение идентификатора черного игрока и его рейтинга.Здесь можно четко рассмотреть имя игрока и его рейтинг, например, у игрока с ником anaisaac рейтинг в процентном соотношении равен 16,3%.

Пункт 5. Диаграмма рассеяния - 2 штуки.

```python
plt.figure(figsize=(7,7))
x = games['moves'][10:15]
y = games['opening_ply'][10:15]
plt.title("Диаграмма рассеяния ходы в стандартной шахматной записи/ходы в начальной фазе")
plt.xlabel('Все ходы в стандартной шахматной записи')
plt.ylabel('Количество ходов в начальной фазе')
plt.scatter(x,y)
plt.show()
```

1. На данной диаграмме рассеяния можно увидеть все ходы в стандартной шахматной записи и количество ходов в начальной фазе. Например, чем меньше у игрока кол-во ходов в стандартной шахматной записи за игру, тем больше ходов в начальной фазе он совершил. Можно рассмотреть, что когда кол-во ходов в стандартной шахматной записи равно чуть минимальному количеству тогда кол-во ходов в начальной фазе тоже равно минимальному, а когда общее количество ходов в стандартной шахматной записи имеет большое колличество, кол-во ходов  в начальной фазе выше среднего.

```python
plt.figure(figsize=(7,7))
x = games['victory_status'].head()
y = games['winner'].head()
plt.title("Диаграмма рассеяния статус игры/победитель")
plt.xlabel('Статус игры')
plt.ylabel('Победитель')
plt.scatter(x,y)
plt.show()
```

2. На данной диаграмме рассеяния можно увидеть статус игры и победителя в этой игре. Статистика показывает что когда цвет игрока победителя белый его статус игры либо outoftime  или mate, а когда цвет игрока победителя черный статус игры resign.

Пункт 6.Гистограмма - 3 штуки.

```python
plt.figure(figsize=(7,7))
plt.hist(games['created_at'], bins = 30, color =["black"],
 edgecolor='pink',alpha=0.7, label ='Распределение времени игроков')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

1. На данной гистограмме можно рассмотреть распределение времени игроков по точным значениям. По результатом гистограммы, понятно, что чем больше времени у игрока уходило на партию, тем ниже у него рейтинг. Например когда рейтинг человека равен 8000, на партию в среднем у него уходило в среднем 1.50  le12 времени.

```python
plt.figure(figsize=(7,7))
plt.hist(games['turns'], bins = 30, color =["pink"],
 edgecolor='black',alpha=0.7, label ='Распределение количество ходов')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

2. На данной гистограмме можно рассмотреть распределение количества ходов по точным значениям. ТОчно видно, что у игроков рейтинга 3000 в среднем колличество ходов равно 50-70.У игроков высокого рейтинга среднее количество ходов 150+.

```python
plt.figure(figsize=(7,7))
plt.hist(games['winner'], bins = 30, color =["pink"],
 edgecolor='black',alpha=0.7, label ='Количество победителей')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

3. На данной гистограмме можно рассмотреть распределение количества победителей по точным значениям.Мы можем видеть цвета победителей и их колличество.Например, победителей белого цвета равно 10000, колличество победителей игроков черного цвета равно около 9000. Ничья в шахматных партиях всего около 1000.
